```python
def dot_product(v1, v2):
    total = 0
    i = 0
    while i < len(v1):
        total = total + v1[i] * v2[i]
        i = i + 1
    return total

a = [1, 2, 3]
b = [4, 5, 6]
print("Dot Product:", dot_product(a, b))
```

```python
import numpy as np

# Create a 3x3 NumPy array
A = np.array([[1, 2, 3],
              [0, 4, 5],
              [1, 0, 6]])

# Transpose
transpose_A = np.transpose(A)

# Determinant
det_A = np.linalg.det(A)

# Inverse
inverse_A = np.linalg.inv(A)

# Print results
print("Original Matrix:\n", A)
print("\nTranspose:\n", transpose_A)
print("\nDeterminant:", det_A)
print("\nInverse:\n", inverse_A)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Dataset
data = {
    'Study_Hours': [2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7],
    'Score':       [21, 47, 27, 75, 30, 20, 88, 60, 81, 25]
}
df = pd.DataFrame(data)

# Features and Target
X = df[['Study_Hours']]  # 2D array
y = df['Score']          # 1D array

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Output
print("Mean Squared Error (MSE):", round(mse, 2))
print("R Score:", round(r2, 2))

# Visualization: Actual vs Predicted
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.scatter(X_test, y_pred, color='red', label='Predicted')
plt.plot(X_test, y_pred, color='green', linestyle='--', label='Regression Line')
plt.xlabel('Study Hours')
plt.ylabel('Score')
plt.title('Actual vs Predicted Scores')
plt.legend()
plt.grid(True)
plt.show()
```
